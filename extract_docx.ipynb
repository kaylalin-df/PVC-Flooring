```python
import zipfile
import xml.etree.ElementTree as ET

docx_path = r"c:\Users\kayla.lin\ai project\PVC Flooring\廠商提供-20251001合作意向書-應昌 (1).docx"

with zipfile.ZipFile(docx_path, 'r') as z:
    with z.open('word/document.xml') as f:
        tree = ET.parse(f)
        root = tree.getroot()

        paragraphs = []
        for para in root.iter('{http://schemas.openxmlformats.org/wordprocessingml/2006/main}p'):
            texts = []
            for node in para.iter('{http://schemas.openxmlformats.org/wordprocessingml/2006/main}t'):
                if node.text:
                    texts.append(node.text)
            paragraphs.append(''.join(texts))

        result = '\n'.join(paragraphs)
        
# Write to a text file so we can read it
output_path = r"c:\Users\kayla.lin\ai project\PVC Flooring\extracted_text.txt"
with open(output_path, 'w', encoding='utf-8') as out:
    out.write(result)
print(result)
```
